```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
from dtw import *
```

```python
P8 = pd.read_csv('C:/Users/baptc/Desktop/Emploi/Formation/WildCodeSchool/Side_project/Data/P8.csv', sep = ';')
P10 = pd.read_csv('C:/Users/baptc/Desktop/Emploi/Formation/WildCodeSchool/Side_project/Data/P10.csv', sep = ';')
```

```python
P8 = np.array(P8['msa'])
P10 = np.array(P10['msa'])
```

```python
plt.figure(figsize = (10, 6))
plt.plot(P8)
plt.plot(P10)
plt.show()
```

```python
distance, path = fastdtw(P8, P10, dist = euclidean)
```

```python
plt.figure(figsize = (15, 7))
for i in range(len(path)):
    plt.plot(path[i][:], [P8[path[i][0]], P10[path[i][1]]], color = 'grey')
    plt.plot(P8, color = 'red')
    plt.plot(P10, color = 'blue')
plt.show()
```

```python
new_P8 = []
new_P10 = []

for i in range(len(path)) :
    new_P8.append(P8[path[i][0]])
    new_P10.append(P10[path[i][1]])
```

```python
plt.figure(figsize = (15, 7))
plt.plot(new_P8)
plt.plot(new_P10)
plt.show()
```
